# Formula da Velocidade Média

```
def calcVelMedia(deltaS, deltaT):
  velMedia = (deltaS / deltaT)
  return velMedia
```

teste de serviço: velocidade media

```
print('Velocidade Média: ',calcVelMedia(50, 4))
```

# Fórmula da aceleração

```
def calcVelMedia(vel, Temp):
  velMedia = (vel / Temp)
  return velMedia
print("aceleração =", calcVelMedia(15, 2))
```

# Fórmulas do Movimento Uniforme (MU)

```
def calcPosicaoMU(posicao_inicial, vel, temp):
  posicao_final = posicao_inicial + (vel * temp)
  return posicao_final
```

## Equação de Torricelli

```
import math

def calcTorricelli(vZero, acel, desloc):
  t = 2 * acel * desloc
  vFinal = t + vZero
  res = round(math.sqrt(vFinal), 2)
  return res
```
